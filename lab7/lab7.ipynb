```python
# Рахматуллин Камиль БВТ2403 Лабораторная работа №7

class Employee():
    def __init__(self, name, id):
        self._name = name
        self._id = id

    def get_info(self):
        return f'Name: {self._name}, Id: {self._id}'

class Manager(Employee):
    def __init__(self, name, id, department):
        self._name = name
        self._id = id
        self._department = department

    def get_info(self):
        return f'Name: {self._name}, Id: {self._id}, Department: {self._department}'

    def manage_project(self):
        print(f'Project in {self._department} managed')

class Technician(Employee):
    def __init__(self, name, id, specialization):
        self._name = name
        self._id = id
        self._specialization = specialization

    def get_info(self):
        return f'Name: {self._name}, Id: {self._id}, Specialization: {self._specialization}'

    def perform_maintenance(self):
        print(f'Maintenance of {self._specialization} performed')

class TechManager(Manager, Technician):
    def __init__(self, name, id, department, specialization):
        Manager.__init__(self, name, id, department)
        Technician.__init__(self, name, id, specialization)
        self.team = []
    def get_info(self):
        return f'Name: {self._name}, Id: {self._id}, Department: {self._department}, Specialization: {self._specialization}'
    def add_employee(self, employee_name, employee_id):
        newEmployee = Employee(employee_name, employee_id)
        self.team.append(newEmployee)
    def get_team_info(self):
        print(self._department, 'team: ')
        for i in self.team:
            print(i.get_info())

employee = Employee('John', 1)
print(employee.get_info())

print(' ')

manager = Manager('Mike', 2, 'Security')
print(manager.get_info())
manager.manage_project()

print(' ')

technician = Technician('Matt', 3, 'Servers')
print(technician.get_info())
technician.perform_maintenance()

print(' ')

techmanager = TechManager('Steve', 4, 'Data Analytics', 'Databases')
print(techmanager.get_info())
techmanager.manage_project()
techmanager.perform_maintenance()
techmanager.add_employee('Mark', 5)
techmanager.add_employee('Carl', 6)
techmanager.get_team_info()







```
